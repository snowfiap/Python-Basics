#                Data cleaning and Dataset creation

```python
# create a dataset with all the types of errors
import numpy as np
student_data={"name":["Snow","Asmi","Kavya","Vennila","Lisha","harishma","Snow"],
              "Roll_no":[52   ,34     ,56     ,   61,        57,    75,         52],
              "age":[20  ,np.nan, 19,        "twenty",      19,      89,        20],
              "Marks":[96    ,35    ,np.nan      ,  97,       48,      np.nan,         96]}
```

```python
student_df=pd.DataFrame(student_data)
student_df
```

```python
student_df.info()
```

### Issues in the student dataframe
    1.Duplicate Values
    2.Error Values
    3.Outlier
    4.Missing values

```python
# Remove duplicate values
student_df.drop_duplicates()
```

```python
#remove duplicate permently
student_df.drop_duplicates(inplace=True)
```

```python
student_df
```

```python
#fix the error calue in age column
student_df["age"][3]=19
```

```python
student_df
```

```python
# fixing outlier
student_df["age"][5]=20
```

```python
student_df
```

```python
# handling Missing values
# 1.removing
# 2.Filling
  
student_df["Marks"][2]=0
student_df["Marks"][4]=0
student_df
```

```python
# To find the mode of age and to fill the age of asmi we need to  convert the datatype
student_df.info()
```

```python
student_df["age"].fillna(int(student_df["age"].mean()),inplace=True)
```

```python
student_df
```

```python
# to fix the missing values in marks replace it with any mark
student_df["Marks"][5]=98
student_df
```

```python
''' to convert the datatype
student_df["age"]=student_df["age"].convert_dtypes(convert_integer=True)
no need to execute because already your column age is already an integer'''
```

### Multivariate Analysis

```python
# consider 3 columns for analysis
store_df.groupby("Region").groups.keys()
```

```python
store_df_west=store_df.groupby("Region").get_group("West")
store_df_west
```

```python
store_df_west.groupby("Category"["sales"])
```

```python
import matplotlib.pyplot as plt
```

```python
# get the data for each group
store_df_west=store_df.groupby("Region").get_group("West")
store_df_central=store_df.groupby("Region").get_group("Central")
store_df_east=store_df.groupby("Region").get_group("East")
store_df_south=store_df.groupby("Region").get_group("South")
```

```python
plt.subplot(2,2,1)
store_df_west.groupby("Category")["Sales"].sum().plot.bar()
plt.subplot(2,2,2)
store_df_central.groupby("Category")["Sales"].sum().plot.bar()
plt.subplot(2,2,3)
store_df_east.groupby("Category")["Sales"].sum().plot.bar()
plt.subplot(2,2,4)
store_df_south.groupby("Category")["Sales"].sum().plot.bar()

```

```python
plt.subplot(2,2,1)
store_df_west.groupby("Category")["Profit"].sum().plot.bar()
plt.subplot(2,2,2)
store_df_central.groupby("Category")["Profit"].sum().plot.bar()
plt.subplot(2,2,3)
store_df_east.groupby("Category")["Profit"].sum().plot.bar()
plt.subplot(2,2,4)
store_df_south.groupby("Category")["Profit"].sum().plot.bar()

```

```python
store_df_Furniture=store_df.groupby("Category").get_group("Furniture")
store_df_OfficeSupplies=store_df.groupby("Category").get_group("Office Supplies")
store_df_Technology=store_df.groupby("Category").get_group("Technology")

```

```python
plt.subplot(2,2,1)
store_df_Furniture.groupby("Ship Mode")["Sales"].sum().plot.bar()
plt.subplot(2,2,2)
store_df_OfficeSupplies.groupby("Ship Mode")["Sales"].sum().plot.bar()
plt.subplot(2,2,3)
store_df_Technology.groupby("Ship Mode")["Sales"].sum().plot.bar()
```
